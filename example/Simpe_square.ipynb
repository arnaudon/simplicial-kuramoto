```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import scipy as sc

from simplicial_kuramoto import SimplicialComplex
from simplicial_kuramoto.graph_generator import modular_graph
from simplicial_kuramoto.integrators import *
from simplicial_kuramoto.plotting import *
```

```python
G=nx.cycle_graph(4)
G.add_edge(0,2)
plt.figure()
nx.draw_networkx(G)
plt.show()

Cff=SimplicialComplex(graph=G) # full full triangles
Cfe=SimplicialComplex(graph=G,faces=[(0,1,2)]) # full empty triangles
Cee=SimplicialComplex(graph=G,faces=[]) # empty empty triangles
```

```python
Ne=Cff.n_edges

# np.random.seed(4444)
initial_phase=2*np.pi*np.random.rand(Ne)
t_max = 1000
n_t = 200

resultff=integrate_edge_kuramoto(Cff, initial_phase, t_max, n_t, omega_0=None)
plot_edge_kuramoto(resultff)
plt.show()
plot_unit_circle(resultff.y)
opff=plot_order_parameter(resultff.y,True,False)
print('Order parameter ff: ',opff[-1])
plot_flow(initial_phase,Cff,resultff,plotname=None)

resultfe=integrate_edge_kuramoto(Cfe, initial_phase, t_max, n_t, omega_0=None)
plot_edge_kuramoto(resultfe)
plt.show()
plot_unit_circle(resultfe.y)
opfe=plot_order_parameter(resultfe.y,True,False)
print('Order parameter fe: ',opfe[-1])
plot_flow(initial_phase,Cfe,resultfe,plotname=None)

resultee=integrate_edge_kuramoto(Cee, initial_phase, t_max, n_t, omega_0=None)
plot_edge_kuramoto(resultee)
plt.show()
plot_unit_circle(resultee.y)
opee=plot_order_parameter(resultee.y,True,False)
print('Order parameter ee: ',opee[-1])
plot_flow(initial_phase,Cee,resultee,plotname=None)
```

```python
ef=np.random.randint(low=0,high=len(Cff.edgelist),size=1)
ef=[2]
print('Flipped edge: ',ef)

initial_phase_flipped=initial_phase.copy()
initial_phase_flipped[ef]=np.mod(np.around(initial_phase[ef]+np.pi, 10), np.around(2 * np.pi, 10))


Cff.flip_edge_orientation(ef[0])
resultff_flipped=integrate_edge_kuramoto(Cff, initial_phase_flipped, t_max, n_t, omega_0=None)
plot_edge_kuramoto(resultff_flipped)
plt.show()
plot_unit_circle(resultff_flipped.y)
opff_flipped=plot_order_parameter(resultff_flipped.y,True,False)
print('Order parameter ff: ',opff_flipped[-1])
plot_flow(initial_phase,Cff,resultff_flipped,plotname=None)

Cfe.flip_edge_orientation(ef[0])
resultfe_flipped=integrate_edge_kuramoto(Cfe, initial_phase_flipped, t_max, n_t, omega_0=None)
plot_edge_kuramoto(resultfe_flipped)
plt.show()
plot_unit_circle(resultfe_flipped.y)
opfe_flipped=plot_order_parameter(resultfe_flipped.y,True,False)
print('Order parameter fe: ',opfe_flipped[-1])
plot_flow(initial_phase,Cfe,resultfe_flipped,plotname=None)

Cee.flip_edge_orientation(ef[0])
resultee_flipped=integrate_edge_kuramoto(Cee, initial_phase_flipped, t_max, n_t, omega_0=None)
plot_edge_kuramoto(resultee_flipped)
plt.show()
plot_unit_circle(resultee_flipped.y)
opee_flipped=plot_order_parameter(resultee_flipped.y,True,False)
print('Order parameter ee: ',opee_flipped[-1])
plot_flow(initial_phase,Cfe,resultee_flipped,plotname=None)
```
